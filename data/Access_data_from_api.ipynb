This notebook explain how the incidence and climate data can be access through the mosqlimate api (https://api.mosqlimate.org/docs/datastore/). 

```python
import numpy as np 
import pandas as pd 
from mosqlient import get_infodengue, get_climate
```

Incidence data:

```python
# return a pd.DataFrame with the data 

df = get_infodengue(

    start_date='2022-12-30', 

    end_date = '2023-01-30', 

    disease = 'dengue',

    uf = 'RJ') # uf represent the state level 

df.head()
```

Climate data: 

```python
# return a pd.DataFrame with the data 
df = get_climate(
    start_date='2022-12-30', 
    end_date = '2023-01-30',
    geocode = 3304557) # geocode represent the municipality level 


df.head()
```
